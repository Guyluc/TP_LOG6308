```python
def ALS(matrix, l_dim, max_iter=100, tolerance=1e-4):
  m, n = matrix.shape
  P = np.zeros((m, l_dim))
  Q = np.random.rand(l_dim, n)

  iter = 0
  dist = float("+inf")
  while iter < max_iter and dist > tolerance:
    iter += 1

    #Estimation de P
    old_P = P.copy()
    P = matrix @ Q.T @ np.linalg.inv(Q @ Q.T)

    #Estimation de Q
    old_Q = Q.copy()
    Q =  np.linalg.inv(P.T @ P) @ P.T @ matrix

    dist = np.linalg.norm(P - old_P) + np.linalg.norm(Q - old_Q)
  
  print("Number of iterations :", iter)
  return P, Q
```

```python
P2, Q2 = ALS(mat, 2)
print("Produit des matrices résultantes :")
print(np.round(P2 @ Q2 ,2))
print('\n\n',"Corrélations obtenues avec les matrices ALS :")
print(np.round(np.corrcoef(P2 @ Q2, rowvar = False), 2))
print('\n',"Corrélations obtenues avec la démo SVD de Landauer :")
print(land_svd2)
```

```python
P3, Q3 = ALS(mat, 3)
print("Produit des matrices résultantes :")
print(np.round(P3 @ Q3 ,2))
print('\n\n',"Corrélations obtenues avec les matrices ALS :")
print(np.round(np.corrcoef(P3 @ Q3, rowvar = False), 2))
print('\n',"Corrélations obtenues avec la démo SVD de Landauer :")
print(land_svd3)
```

Les résultats obtenus avec les matrices déterminées par ALS sont quasiment identiques à ceux obtenus par SVD. A 2 dimensions on retrouve une corrélation parfaite entre les articles de la convention d'algèbre linéaire et assez bonne pour la plupart des cas de la convention d'IHM mais assez mauvais pour les articles 2 et 5. De même, les résultats sont un peu moins bon avec 3 dimensions comme avec SVD car le problème n'a que deux classes donc le 3 axes n'est pas intéressant.

#3. Utilisation du temps

Dans une approche par factorisation de matrice on peut définir des biais qui varient en fonction du temps pour la prédiction du vote de l'utilisateur u pour l'item i par rapport à la moyenne des votes μ : 

$$ \hat{r}_{ui}(t) = μ + b_i(t) + b_u(t) + q^T_ip_u(t) $$

$ b_i(t) $ représente le biais d'un item, son écart de valeur par rapport à la moyenne, il est par nature assez fixe et peut varier autour de sa moyenne selon la période (comme la saison dans l'année) :

$$ b_i(t) = b_i + b_{i,period}(t) $$

$ b_u(t) $ représente le biais d'un utilisateur et on pourrait le faire varier dans le temps selon l'éloignement par rapport à la date moyenne de vote ($t_u$) :

$$ b_u (t) = b_u + α_u · dev_u(t) $$

$$ dev_u(t) = sign(t − t_u) · |t − t_u|^β $$

Enfin, les matrices P et Q n'apprenent que les variations propres à chaque utilisateur et item par rapport à ces moyennes. On peut aussi modéliser dans la matrice P des variations linéaires de l'utilisateur par rapport à sa propre moyenne selon les jours qui passent pour chaque item spécifiquement qui représentera plus son changement de goûts :

$$ p_{uk}(t) = p_{uk} + α_{uk} · dev_u(t) + p_{uk,t}\ \ \ \ \ \ \ \ k = 1,...,f $$
